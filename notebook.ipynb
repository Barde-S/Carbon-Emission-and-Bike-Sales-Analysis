## üíæ The data I

### You have access to seven years of CO2 emissions data for Canadian vehicles ([source](https://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64#wb-auto-6)):

- "Make" - The company that manufactures the vehicle.
- "Model" - The vehicle's model.
- "Vehicle Class" - Vehicle class by utility, capacity, and weight.
- "Engine Size(L)" - The engine's displacement in liters.
- "Cylinders" - The number of cylinders.
- "Transmission" - The transmission type: A = Automatic, AM = Automatic Manual, AS = Automatic with select shift, AV = Continuously variable, M = Manual, 3 - 10 = the number of gears.

- "Fuel Consumption Comb (L/100 km)" - Combined city/highway (55%/45%) fuel consumption in liters per 100 km (L/100 km).
- "CO2 Emissions(g/km)" - The tailpipe carbon dioxide emissions in grams per kilometer for combined city and highway driving. 

The data comes from the Government of Canada's open data [website](https://open.canada.ca/en).

### 1b. The least expensive?


```python
SELECT product_name, list_price
FROM products
WHERE list_price = (SELECT MIN(list_price) FROM products);
```

The least expensive item soldby thr company is Strider Classic 12 Balance Bike - 2018, with price of $89.99

### 2. How many different products of each category does your company sell?

```python
SELECT distinct(category_name), COUNT(*) AS num_products
FROM products
JOIN categories
ON products.category_id = categories.category_id
GROUP BY products.category_id, categories.category_name;
```

The company sell's 59 different products for children bicycles category, 30 in comfort bicycles, 78 cruiser bicycles, 10 cyclocross bicycles, 24 of electric bikes, 60 for both moutain and road bike.

### 3. What are the top three brands with the highest average list price? The top three categories?

```python
SELECT TOP 3 brand_name, AVG(list_price) AS avg_price
FROM products
JOIN brands
ON products.brand_id = brands.brand_id
GROUP BY products.brand_id, brands.brand_name
ORDER BY avg_price DESC;
```

The top three brands with the highest list price are Trek ($2500.06), Heller($2172.99), and Surly ($1331.75
) respectively.

### Other insights

### What is the distribution of products across categories? (Are there any categories with significantly fewer or more products than others?)

```python
SELECT categories.category_name, COUNT(products.product_id) AS num_products, 
ROUND(COUNT(products.product_id)*100.0/SUM(COUNT(products.product_id)) OVER(),2) AS percentage_of_products
FROM products
JOIN categories
ON products.category_id = categories.category_id
GROUP BY categories.category_name
ORDER BY num_products DESC;
```

Cyclocross Bicycles, Electric Bikes and Comfort Bicycles are  categories with few products

### Are there any brands or categories with significantly higher or lower prices than others?

```python
SELECT brands.brand_name, categories.category_name, AVG(products.list_price) AS avg_price
FROM products
JOIN brands
ON products.brand_id = brands.brand_id
JOIN categories
ON products.category_id = categories.category_id
GROUP BY brands.brand_name, categories.category_name
ORDER BY avg_price DESC;
```

Trek and electra are the brand with significantly higher price base on brand and category while children bicycle is the category with significantly lower price.

### Average list price for product categories

```python
SELECT category_name, AVG(list_price) AS avg_price
FROM products
JOIN categories
ON products.category_id = categories.category_id
GROUP BY products.category_id, categories.category_name
ORDER BY avg_price DESC
```

The list of product categories ranks electric bikes as the most expensive, followed by road bikes, and finally cyclocross bikes, with their average prices shown above.
