### You have access to seven years of CO2 emissions data for Canadian vehicles ([source](https://open.canada.ca/data/en/dataset/98f1a129-f628-4ce4-b24d-6f16bf24dd64#wb-auto-6)):

- "Fuel Consumption Comb (L/100 km)" - Combined city/highway (55%/45%) fuel consumption in liters per 100 km (L/100 km).
- "CO2 Emissions(g/km)" - The tailpipe carbon dioxide emissions in grams per kilometer for combined city and highway driving. 

The data comes from the Government of Canada's open data [website](https://open.canada.ca/en).

### 1. What is the median engine size in liters?

```python
median_engine_size = cars['Engine Size(L)'].median()
print(f'The median engine size is: { median_engine_size} liters')
```

### 2. What is the average fuel consumption for regular gasoline (Fuel Type = X), premium gasoline (Z), ethanol (E), and diesel (D)?

```python
fuel_types = ['D', 'E', 'X', 'Z']
for fuel_type in fuel_types:
    avg_fuel_consumption = cars[cars['Fuel Type'] == fuel_type]['Fuel Consumption Comb (L/100 km)'].mean()
    print(f'The average fuel consumption for, {fuel_type} is {avg_fuel_consumption}, Litre/100 km.')
```

### 3. What is the correlation between fuel consumption and CO2 emissions?

```python
correlation = cars['Fuel Consumption Comb (L/100 km)'].corr(cars['CO2 Emissions(g/km)'])
print(f'The correlation between fuel consumption and CO2 emissions is {correlation}')
```

From above, one can see that there is a very strong correlation between CO2 emission and fuel consumption. This means that the more the fuel consumption the more the combustion (CO2 emission)

### 4. Which vehicle class has lower average CO2 emissions, 'SUV - SMALL' or 'MID-SIZE'?

```python
co2_by_class = (cars.groupby('Vehicle Class')['CO2 Emissions(g/km)'].mean())
print(co2_by_class)
print('------------'*5)
if co2_by_class['SUV - SMALL'] < co2_by_class['MID-SIZE']:
    print('SUV - SMALL has lower average CO2 emissions.')
else:
    print('MID-SIZE has lower average CO2 emissions.')
```

Mid-size vehicles have an average of 222.455428 CO2 emission, while suv-smaall average 236.292523 CO2 emission. if these figures are compared, can clearly see that suv-smaal has a higher average of CO2 emission out of the two.

### 5. What are the average CO2 emissions for all vehicles? For vehicles with an engine size of 2.0 liters or smaller?

```python
avg_co2_all = cars['CO2 Emissions(g/km)'].mean()
avg_co2_small_engine = cars[cars['Engine Size(L)'] <= 2.0]['CO2 Emissions(g/km)'].mean()
display(f'The average CO2 emissions for all vehicles is {avg_co2_all} g/km.')
display(f'The average CO2 emissions for vehicles with an engine size of 2.0 liters or smaller is, {avg_co2_small_engine} g/km.')
```
